```python
from requests.exceptions import RequestException
from bs4 import BeautifulSoup
import requests
import re
import time
from multiprocessing import Pool
import csv
import threading
```

```python
def get_one_page(url):
    try:
        header ={
            'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.96 Safari/537.36'
        }
        response = requests.get(url,headers=header)
        if response.status_code == 200:
            return response.text
        return None
    except RequestException:
        return None

    
def parse_one_page(html,offset):
#     result = ['id','name','man','woman','below20','20~24','25~29','30~34','35~39','above40','bachelor degree or above',
#               'undergraduate and below','second-tier city','fourth-tier city','third-tier citiy','first-tier city','occupation','',''
#              'preference-film','','','','','bought']
    result = []
    soup = BeautifulSoup(str(html),'lxml')
    result.append(offset)
    
    name = soup.select('#pageData')
    name = re.findall('movieName":"(.*?)"',str(name),re.S)
    result.append(str(name).replace("['","").replace("']",""))
    
    for sex in soup.select('.stackcolumn-bar')[:2]:
        result.append(round(float(sex.attrs['style'].replace("width: ","").replace("%",""))/100,2))
    
    for age in soup.select('.bar-group g'):
        result.append(round(float(age.get_text().replace("%",""))/100,2))
    
    for edu in soup.select('.stackcolumn-bar')[2:]:
        result.append(round(float(edu.attrs['style'].replace("width: ","").replace("%",""))/100,2))
   
    for area in soup.select('.linebars-value'):
        result.append(round(float(area.get_text().replace("%",""))/100,2))
    
    for occupation in soup.select('.pie-label-line text')[:1]: 
        result.append((occupation.get_text()[:2]))
        result.append(round(float(occupation.get_text()[2:].replace("%",""))/100,2))
    
    for preference in soup.select('.pie-label-line text')[1:]:
        result.append(preference.get_text()[:2])
        result.append(round(float(preference.get_text()[2:].replace("%",""))/100,2))
    
    for bought in soup.select('.scroll-title'):
        result.append(bought.get_text())
    print(result)
    return result



def writein(result):
    #打开文件，追加a
    try:
        out = open('yonghuhuaxiang3.csv','a', newline='')
        #设定写入模式
        csv_write = csv.writer(out,dialect = 'excel')
        #写入具体内容
        csv_write.writerow(result)
        print ("write over")
    except UnicodeError:
        print("go on")

def main(offset):
    url = 'https://piaofang.maoyan.com/movie/'+str(offset)+'/wantindex?'
    html = get_one_page(url)
    result = parse_one_page(html,offset)
    writein(result)


def readin():
    for line in open("2016id.txt"):
        main(line.replace("\n",""))
    # for i in range(0,1):
#         main(0)
if __name__ == '__main__':
    for i in range(1000):
        t = threading.Thread(target=readin())
        thread_list.append(t)
    for t in thread_list:
        t.start()
```
