```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
url = 'https://piaofang.maoyan.com/movie/410629/wantindex?'#输入想要爬取的网址（其实只需要改一下movie/后边的数字，每个电影有自己专属的数字号）
header ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.96 Safari/537.36'}
response = requests.get(url,headers=header)
soup = BeautifulSoup(str(response.text),'lxml')
print(soup.prettify())#格式化输出
```

```python
#受众性别占比
for text in soup.select('h2')[:1]:
    print(text.get_text())
for text in soup.select('.stackcolumn-desc-text')[:2]:
    print(text.get_text())
for sex in soup.select('.stackcolumn-bar')[:2]:
    print(sex.attrs['style'])
print('\n')

#受众年龄占比
for text in soup.select('h2')[1:2]:
    print(text.get_text())
for text in soup.select('.x g'):
    print(text.get_text())
for text in soup.select('.bar-group g'):
    print(text.get_text())
print('\n')

#教育程度
for text in soup.select('h2')[2:3]:
    print(text.get_text())
for text in soup.select('.stackcolumn-desc-text')[2:]:
    print(text.get_text()) 
for sex in soup.select('.stackcolumn-bar')[2:]:
    print(sex.attrs['style'])
print('\n')

#活跃地区
for text in soup.select('h2')[3:4]:
    print(text.get_text())
for text in soup.select('.linebars-content'):
    print(text.get_text())
print('\n')
    
#职业
for text in soup.select('h2')[4:5]:
    print(text.get_text())
for text in soup.select('.pie-label-line')[:1]:
    print(text.get_text())
print('\n')
    
#电影类型偏好    
for text in soup.select('h2')[5:6]:
    print(text.get_text())
for text in soup.select('.pie-label-line')[1:]:
    print(text.get_text())
print('\n')
    
#他们买过
for text in soup.select('h2')[6:]:
    print(text.get_text())
for text in soup.select('.scroll'):
    print(text.get_text())
```
